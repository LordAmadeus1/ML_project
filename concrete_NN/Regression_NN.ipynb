# Regression Neuroal Network with Keras



import all the necessary libraries

```python
import pandas as pd
import keras
import sklearn
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

# Download and Clean the Dataset

Knowing that the data set is in the link https://cocl.us/concrete_data.


```python
url = 'https://cocl.us/concrete_data'
```

```python
df = pd.read_csv(url)
df.head(3)
```

```python
df.columns.to_list()
```

```python
df.shape #number of points
```

```python
df.describe()
```

```python
df.isnull().sum() #number of null values per column
```

#Creating X, Y and spliting

```python
X = df.drop(columns='Strength')  # Caracteristics
y = df['Strength'] #target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
#We specify random_state=42 for reproduce the results
```

```python
n_cols = X_train.shape[1]
```

#Build a Neuronal Network

```python
def regression_nn_model():
  model = Sequential() #Initialize the model

  model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
  model.add(Dense(1))

  model.compile(optimizer='adam', loss='mean_squared_error') #compile the model
  return model
```

Train and Test the Network

```python
model = regression_nn_model()
```

```python
model.fit(X_train, y_train,validation_split=0.3, epochs=50, verbose=2 )
```

#Model Evaluation

```python
loss = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)

loss
```

```python
mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
std = np.std(mse)

mean, std
```

The mean_squared_error doesn't look well

Let's try to make it better.
What if we normilize our data?

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y,test_size=0.30, random_state=None)
```

```python
X_train.shape, y_train.shape
```

```python
model = regression_nn_model()

model.fit(X_train, y_train, validation_split=0.2 ,epochs=50, verbose=2 )
```

```python
loss = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)

loss
```

```python
mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
std = np.std(mse)

mean, std
```

Okey, these look worst than before.
We need to try to make it better

#More models and splits?

```python
mse_list = []
#First is just an empty list

for i in range(50):
  #we try new seeds
  X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3,random_state=None)
  model.fit(X_train, y_train, epochs=50, verbose=2)
  y_pred = model.predict(X_test)
  mse= mean_squared_error(y_test, y_pred)

  mse_list.append(mse)
```

We change our list to array cause is  easy to work with

```python
mse_array = np.array(mse_list)
```

```python

mse_mean = mse_array.mean()
#These will return us the mean of the mean_squared_error of all the epochs

mse_std = mse_array.std()
#The standard desviation

mse_mean, mse_std
```

Nice much better, at least the mse

#What if we increase the number of hidden layers?

We need to build another Neuronal Network or just make better that we have

```python
def regression_nn_model():
  model = Sequential()

  model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(10, activation='relu'))
  model.add(Dense(1))

  model.compile(optimizer='adam', loss='mean_squared_error')
  return model
```

```python
model = regression_nn_model()
```

```python
model.fit(X_train, y_train, epochs=50, verbose=2)
```

```python
loss = model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)

loss
```

```python
mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
std = np.std(mse)

mean, std
```

Much better, what if we add more epochs?


```python
model.fit(X_train, y_train, validation_split=0.2 ,epochs=100, verbose=2)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mean = np.mean(mse)
std = np.std(mse)

mean, std
```
